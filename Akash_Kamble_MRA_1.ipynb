# Problem 1:

### Uploading necessary libraries

```python
import numpy as np   
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt   

import warnings
warnings.filterwarnings("ignore")
```

### Importing the data

```python
df = pd.read_excel('Sales_Data.xlsx',sheet_name='Sheet1')
df.head(10)
```

### EDA

```python
# Shape of the data
print('No. of rows: {}'.format(df.shape[0]))
print('No. of colums: {}'.format(df.shape[1]))
```

```python
# Data types
df.info()
```

```python
# Ploting the boxplots
plt.figure(figsize=(15,20))
cont=['QUANTITYORDERED','PRICEEACH', 'SALES', 'DAYS_SINCE_LASTORDER', 'MSRP']
for i in range(len(cont)):
    plt.subplot(3,2,i+1)
    sns.boxplot(y=df[cont[i]], data = df)
    plt.title('Boxplot for {}'.format(cont[i]))
    plt.tight_layout(pad=3)
```

```python
# Ploting the distributions
plt.figure(figsize=(15,12))
cont=['QUANTITYORDERED','PRICEEACH', 'SALES', 'DAYS_SINCE_LASTORDER', 'MSRP']
for i in range(len(cont)):
    plt.subplot(3,2,i+1)
    sns.distplot(df[cont[i]])
    plt.title('Boxplot for {}'.format(cont[i]))
    plt.tight_layout(pad=3)
```

```python
# Cheking the variables

for column in df.columns:
    print(column.upper(), ': ', df[column].nunique())
    print(df[column].value_counts().sort_values())
```

```python
# Descibe the variables
df.describe(include='all').T
```

```python
# Checking got null values
df.isnull().sum()
```

```python
# Checking for the duplicate rows
dups = df.duplicated()
print('No. of duplicated rows: {}'.format(dups.sum()))
```

### Univariate Analysis

```python
fig, axes = plt.subplots(nrows=5,ncols=2)
fig.set_size_inches(7,20)
a = sns.distplot(df['QUANTITYORDERED'] , ax=axes[0][0])
a.set_title("QUANTITYORDERED Distribution",fontsize=10)

a = sns.boxplot(df['QUANTITYORDERED'] , orient = "v" , ax=axes[0][1])
a.set_title("QUANTITYORDERED",fontsize=10)

a = sns.distplot(df['PRICEEACH'] , ax=axes[1][0])
a.set_title("PRICEEACH",fontsize=10)

a = sns.boxplot(df['PRICEEACH'] , orient = "v" , ax=axes[1][1])
a.set_title("PRICEEACH",fontsize=10)


a = sns.distplot(df['SALES'] , ax=axes[2][0])
a.set_title("SALES",fontsize=10)

a = sns.boxplot(df['SALES'] , orient = "v" , ax=axes[2][1])
a.set_title("SALES",fontsize=10)


a = sns.distplot(df['DAYS_SINCE_LASTORDER'] , ax=axes[3][0])
a.set_title("DAYS_SINCE_LASTORDER",fontsize=10)

a = sns.boxplot(df['DAYS_SINCE_LASTORDER'] , orient = "v" , ax=axes[3][1])
a.set_title("DAYS_SINCE_LASTORDER",fontsize=10)

a = sns.distplot(df['MSRP'] , ax=axes[4][0])
a.set_title("MSRP",fontsize=10)

a = sns.boxplot(df['MSRP'] , orient = "v" , ax=axes[4][1])
a.set_title("MSRP",fontsize=10)

plt.tight_layout()

plt.show()
```
